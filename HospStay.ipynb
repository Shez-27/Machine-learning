```python
StarWars = pd.read_csv(StarWars_p, sep=',', decimal='.', encoding = 'latin1', header=1, names=['RespondentID','Seen any of the 6 films in the Star Wars ?', 
                                                                             'Do you consider fan of the Star Wars film ?','Seen episode 1',
                                                                             'Seen episode 2','Seen episode 3','Seen episode 4','Seen episode 5',
                                                                             'Seen episode 6','Rank for episode 1','Rank for episode 2',
                                                                             'Rank for episode 3','Rank for episode 4','Rank for episode 5',
                                                                             'Rank for episode 6','Han Solo character view','Luke Skywalker character view',
                                                                             'Princess Leia Organa character view','Anakin Skywalker character view',
                                                                             'Obi Wan Kenobi character view','Emperor Palpatine character view',
                                                                             'Darth Vader character view','Lando Calrissian character view','Boba Fett character view',
                                                                             'C-3P0 character view','R2 D2 character view','Jar Jar Binks character view',
                                                                             'Padme Amidala character view','Yoda character view','Character shot first',
                                                                             'Familiar with the Expanded Universe','Consider yourself to be a fan of the Expanded Universe',
                                                                             'Consider yourself to be a fan of the Star Trek','Gender','Age','Household Income','Education',
                                                                             'Location (Census Region)'])
```

```python
#1.2 To check  the data types
StarWars.dtypes
```

```python
columns= list(StarWars)
```

```python
#1.3 To check typos
#Used for loop to go through every columns across the rows and value_counts() to get the counts of each value inside a column
for i in columns:
    temp = StarWars[i].value_counts()
    print(temp)
    print("\r\n")
```

```python
#1.3 To correct the typos
mask_fan_of_film_Yess=(StarWars['Do you consider fan of the Star Wars film ?']== 'Yess')
StarWars.loc[mask_fan_of_film_Yess, 'Do you consider fan of the Star Wars film ?']= 'Yes'
mask_fan_of_film_Noo=(StarWars['Do you consider fan of the Star Wars film ?']== 'Noo')
StarWars.loc[mask_fan_of_film_Noo, 'Do you consider fan of the Star Wars film ?']= 'No'
```

```python
mask_fan_of_exp_universe_Yess=(StarWars['Consider yourself to be a fan of the Expanded Universe']== 'Yess')
StarWars.loc[mask_fan_of_exp_universe_Yess, 'Consider yourself to be a fan of the Expanded Universe']= 'Yes'
```

```python
mask_fan_of_star_trek_yes=(StarWars['Consider yourself to be a fan of the Star Trek']== 'yes')
StarWars.loc[mask_fan_of_star_trek_yes, 'Consider yourself to be a fan of the Star Trek']= 'Yes'
mask_fan_of_star_trek_no=(StarWars['Consider yourself to be a fan of the Star Trek']== 'no')
StarWars.loc[mask_fan_of_star_trek_no, 'Consider yourself to be a fan of the Star Trek']= 'No'
mask_fan_of_star_trek_Noo=(StarWars['Consider yourself to be a fan of the Star Trek']== 'Noo')
StarWars.loc[mask_fan_of_star_trek_Noo, 'Consider yourself to be a fan of the Star Trek']= 'No'
```

```python
mask_gender_female=(StarWars['Gender']== 'female')
StarWars.loc[mask_gender_female, 'Gender']= 'Female'
mask_gender_male=(StarWars['Gender']== 'male')
StarWars.loc[mask_gender_male, 'Gender']= 'Male'
mask_gender_F=(StarWars['Gender']== 'F')
StarWars.loc[mask_gender_F, 'Gender']= 'Female'
```

```python
#To find the value count of all the values inside each column
for i in columns:
    temp = StarWars[i].value_counts()
    print(temp)
    print("\r\n")
```

```python
#1.4 To eliminate extra white spaces
columns = list(StarWars)
for i in columns:
    if StarWars[i].dtypes == object:
        StarWars[i] = StarWars[i].str.strip()
```

```python
#1.5 To converte all the data in the dataset to upper case
for columns in StarWars.columns:
    if(StarWars[columns].dtype == object ):
        StarWars[columns]=StarWars[columns].str.upper()
```
